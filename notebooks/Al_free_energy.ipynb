# Notebook for wrapper functions that run calphy for given purpose

```python
import calphy_interface as ci
```

## Structure and potential

```python
from ase.build import bulk

Al_pure_structure = bulk('Al', cubic=True).repeat(3)
```

```python
from pyiron_lammps import get_potential_by_name

potential_df = get_potential_by_name('2009--Mendelev-M-I--Al-Mg--LAMMPS--ipr1')
potential_df= potential_df.to_frame().transpose()
potential_df
```

## Running calphy

### Free energy at a particular temperature

```python
input_params_fe = {  
    "mode": "fe", 
    "temperature": 300,  
    "n_equilibration_steps": 2500,
    "n_switching_steps": 2500,
    "n_print_steps": 1000,
    "equilibration_control": "nose-hoover",
    "queue": {
        "cores": 1,  
        "scheduler": "local"  
    },
    'reference_phase': 'solid',
    'file_format': 'lammps-data',
} 
```

```python
calphy_results_fe = ci.calc_free_energy_with_calphy(
    input_structure=Al_pure_structure,
    potential_df=potential_df,
    calphy_parameters=input_params_fe,
    working_directory='Aluminium_free_energy/Al_fe',
    user_dict={}
)
```

```python
calphy_results_fe[0]
```

```python
calphy_results_fe[1]
```
